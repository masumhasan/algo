```python
import matplotlib.pyplot as plt
import numpy as np

# Data points
number_of_elements = [100, 500, 1000, 5000, 10000]
time_taken = [0.000009, 0.000037, 0.000083, 0.000529, 0.000873]
total_comparisons = [550, 3855, 8708, 55214, 120529]

# Plotting time taken versus number of elements
plt.figure(1)
plt.plot(number_of_elements, time_taken, 'bo', label='Data Points')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time Taken (seconds)')
plt.title('Time Taken versus Number of Elements')

# Adding linear regression line
m, b = np.polyfit(number_of_elements, time_taken, 1)
plt.plot(number_of_elements, m*np.array(number_of_elements) + b, 'r--', label='Linear Regression')

# Display the legend
plt.legend()

# Plotting total comparisons versus number of elements
plt.figure(2)
plt.plot(number_of_elements, total_comparisons, 'go', label='Data Points')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Total Comparisons')
plt.title('Total Comparisons versus Number of Elements')

# Adding linear regression line
m, b = np.polyfit(number_of_elements, total_comparisons, 1)
plt.plot(number_of_elements, m*np.array(number_of_elements) + b, 'r--', label='Linear Regression')

# Display the legend
plt.legend()

# Display the plots
plt.show()

```

```python
import plotly.graph_objects as go

# Data
elements = [100, 500, 1000, 5000, 10000]
time_taken = [0.000009, 0.000037, 0.000083, 0.000529, 0.000873]
total_comparisons = [550, 3855, 8708, 55214, 120529]

# Linear regression
time_regression = np.polyfit(elements, time_taken, 1)
comparisons_regression = np.polyfit(elements, total_comparisons, 1)

# Regression lines
time_regression_line = np.polyval(time_regression, elements)
comparisons_regression_line = np.polyval(comparisons_regression, elements)

# Plotting
fig = go.Figure()

# Time taken graph
fig.add_trace(go.Scatter(x=elements, y=time_taken, mode='markers', name='Time Taken'))
fig.add_trace(go.Scatter(x=elements, y=time_regression_line, mode='lines', name='Time Regression'))

# Total comparisons graph
fig.add_trace(go.Scatter(x=elements, y=total_comparisons, mode='markers', name='Total Comparisons'))
fig.add_trace(go.Scatter(x=elements, y=comparisons_regression_line, mode='lines', name='Comparisons Regression'))

# Layout
fig.update_layout(
    title="Linear Regression Graph",
    xaxis_title="Number of Elements",
    yaxis_title="Time Taken / Total Comparisons"
)

# Save the plot as HTML
fig.write_html('linear_regression_graph.html')
```

```python
import plotly.graph_objects as go

# Data
elements = [100, 500, 1000, 5000, 10000]
time_taken = [0.000009, 0.000037, 0.000083, 0.000529, 0.000873]
total_comparisons = [550, 3855, 8708, 55214, 120529]

# Linear regression
time_regression = np.polyfit(elements, time_taken, 1)
comparisons_regression = np.polyfit(elements, total_comparisons, 1)

# Regression lines
time_regression_line = np.polyval(time_regression, elements)
comparisons_regression_line = np.polyval(comparisons_regression, elements)

# Figure 1: Time taken versus Number of elements
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=elements, y=time_taken, mode='markers', name='Time Taken'))
fig1.add_trace(go.Scatter(x=elements, y=time_regression_line, mode='lines', name='Time Regression'))
fig1.update_layout(
    title="Time Taken vs Number of Elements",
    xaxis_title="Number of Elements",
    yaxis_title="Time Taken"
)
fig1.write_html('time_taken_graph.html')

# Figure 2: Total comparisons versus Number of elements
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=elements, y=total_comparisons, mode='markers', name='Total Comparisons'))
fig2.add_trace(go.Scatter(x=elements, y=comparisons_regression_line, mode='lines', name='Comparisons Regression'))
fig2.update_layout(
    title="Total Comparisons vs Number of Elements",
    xaxis_title="Number of Elements",
    yaxis_title="Total Comparisons"
)
fig2.write_html('total_comparisons_graph.html')
```
